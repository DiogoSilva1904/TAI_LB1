# Entropy Variation

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("entropy_data.csv")


plt.plot(data["position"], data["entropy_value"], label="Entropy")
plt.ylim(0, 4)  # Limits entropy values between 0 and 5
plt.xlim(data["position"].min(), data["position"].max())  # Auto scale x-axis

plt.xlabel("Position in Text")
plt.ylabel("Entropy")
plt.title("Entropy Variation in Text")
plt.legend()

plt.show()
```
